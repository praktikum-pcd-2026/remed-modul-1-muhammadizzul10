```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

Baris kode di atas digunakan untuk mengimpor tiga library penting dalam pengolahan data dan citra di Python, yaitu NumPy untuk mengelola dan memproses data dalam bentuk array atau matriks, OpenCV (cv2) untuk membaca serta mengolah gambar atau video, dan Matplotlib (pyplot) untuk menampilkan hasil pengolahan dalam bentuk gambar atau grafik sehingga memudahkan visualisasi.

```python
citra1 = cv.imread('Assets/citra1.PNG', cv.IMREAD_GRAYSCALE)
citra2 = cv.imread('Assets/citra2.PNG', cv.IMREAD_GRAYSCALE)

plt.subplot(1, 2, 1)
plt.imshow(citra1, cmap = 'gray')

plt.subplot(1, 2, 2)
plt.imshow(citra2, cmap = 'gray')
```

Baris kode di atas digunakan untuk membaca dua gambar langsung dalam format grayscale menggunakan OpenCV, kemudian menampilkannya berdampingan menggunakan Matplotlib dengan susunan 1 baris 2 kolom agar kedua citra dapat dibandingkan secara visual.

```python
def rotate(image, val): 
    h = image.shape[0]
    w = image.shape[1] 
    val = int(val) % 360  
    if val == 0: return image 
    elif val == 90: 
        rot = np.zeros((w, h), dtype=image.dtype) 
        for i in range(h): 
            for j in range(w): rot[j, h - 1 - i] = image[i, j] 
        return rot 
    elif val == 180: 
        rot = np.zeros_like(image, dtype=image.dtype) 
        for i in range(h): 
            for j in range(w): rot[h - 1 - i, w - 1 - j] = image[i, j] 
        return rot 
    elif val == 270:  
        rot = np.zeros((w, h), dtype=image.dtype) 
        for i in range(h): 
            for j in range(w): rot[w - 1 - j, i] = image[i, j] 
        return rot 
    return image 
```

Baris kode di atas mendefinisikan fungsi rotate yang digunakan untuk memutar citra berdasarkan sudut tertentu (0°, 90°, 180°, atau 270°), dengan cara mengatur ulang posisi piksel menggunakan perulangan dan indeks array, sehingga menghasilkan citra baru sesuai arah rotasi yang diinginkan.

```python
rotasiCitra1 = rotate(citra1, val = 180)

plt.imshow(rotasiCitra1, cmap = 'gray')
plt.show()
```

Baris kode di atas digunakan untuk memutar citra citra1 sebesar 180 derajat menggunakan fungsi rotate, kemudian hasilnya disimpan dalam variabel rotasiCitra1 dan ditampilkan menggunakan Matplotlib dalam bentuk grayscale.

```python
def gabung(image1, image2):
    merged = image1 + image2;
    return merged;
```

Baris kode di atas mendefinisikan fungsi gabung yang digunakan untuk menggabungkan dua citra dengan cara menjumlahkan nilai piksel dari image1 dan image2, sehingga menghasilkan citra baru yang merupakan kombinasi intensitas dari kedua gambar tersebut.

```python
citraGabung = gabung(rotasiCitra1, citra2)

plt.imshow(citraGabung, cmap = 'gray')
```

Baris kode di atas digunakan untuk menggabungkan citra hasil rotasi (rotasiCitra1) dengan citra kedua (citra2) menggunakan fungsi gabung, kemudian hasilnya disimpan dalam variabel citraGabung dan ditampilkan menggunakan Matplotlib dalam bentuk grayscale.

```python
def potong(gambar, a, b, mode='H'): 
    return gambar[a:b, :] if mode == 'H' else gambar[:, a:b]
```

Baris kode di atas mendefinisikan fungsi potong yang digunakan untuk memotong citra berdasarkan rentang indeks tertentu, di mana jika mode 'H' maka pemotongan dilakukan secara horizontal (baris dari a sampai b), sedangkan selain itu dilakukan secara vertikal (kolom dari a sampai b).

```python
citraPotong1 = potong(citraGabung, 0, 390, mode='H')
citraPotong2 = potong(citraGabung, 390, 780, mode='H')

plt.imshow(citraPotong1, cmap = 'gray')
plt.show()

plt.imshow(citraPotong2, cmap = 'gray')
plt.show()
```

Baris kode di atas digunakan untuk memotong citra gabungan (citraGabung) menjadi dua bagian secara horizontal berdasarkan rentang baris tertentu, kemudian masing-masing hasil potongan disimpan dalam citraPotong1 dan citraPotong2, lalu ditampilkan satu per satu menggunakan Matplotlib dalam bentuk grayscale.

```python
citraPotong1_v1 = potong(citraPotong1, 0, 390, mode='V')
citraPotong1_v2 = potong(citraPotong1, 390, 780, mode='V')
citraPotong2_v1 = potong(citraPotong2, 0, 390, mode='V')
citraPotong2_v2 = potong(citraPotong2, 390, 780, mode='V')

plt.figure(figsize=(7,7))

plt.subplot(2, 2, 1)
plt.imshow(citraPotong1_v1, cmap = 'gray')
plt.title('Citra 1')

plt.subplot(2, 2, 2)
plt.imshow(citraPotong1_v2, cmap = 'gray')
plt.title('Citra 2')

plt.subplot(2, 2, 3)
plt.imshow(citraPotong2_v1, cmap = 'gray')
plt.title('Citra 3')

plt.subplot(2, 2, 4)
plt.imshow(citraPotong2_v2, cmap = 'gray')
plt.title('Citra 4')
```

Baris kode di atas digunakan untuk memotong kembali dua citra sebelumnya secara vertikal menjadi empat bagian berbeda, kemudian masing-masing hasil potongan ditampilkan dalam satu tampilan dengan susunan 2x2 menggunakan Matplotlib dan diberi judul untuk setiap citra.

```python
def flip(image, mode='X'): 
    h, w = image.shape
    hasil = np.zeros((h, w), dtype=image.dtype) 
    for i in range(h): 
        for j in range(w): 
            if mode == 'X': 
                hasil[i, j] = image[i, w - 1 - j] 
            elif mode == 'Y': 
                hasil[i, j] = image[h - 1 - i, j] 
    return hasil
```

Baris kode di atas mendefinisikan fungsi flip yang digunakan untuk membalik citra berdasarkan sumbu tertentu, di mana jika mode 'X' maka gambar dibalik secara horizontal (kiri ke kanan) dan jika mode 'Y' maka dibalik secara vertikal (atas ke bawah), dengan cara menukar posisi piksel menggunakan perulangan.

```python
citra1Benar = rotate(citraPotong1_v1, 270)
citra2Benar = flip(citraPotong1_v2, mode='X')
citra3Benar = rotate(citraPotong2_v1, 270)
citra4Benar = flip(citraPotong2_v2, mode='X')

plt.figure(figsize=(7,7))

plt.subplot(2, 2, 1)
plt.imshow(citra1Benar, cmap = 'gray')
plt.title('Citra 1')

plt.subplot(2, 2, 2)
plt.imshow(citra4Benar, cmap = 'gray')
plt.title('Citra 2')

plt.subplot(2, 2, 3)
plt.imshow(citra2Benar, cmap = 'gray')
plt.title('Citra 3')

plt.subplot(2, 2, 4)
plt.imshow(citra3Benar, cmap = 'gray')
plt.title('Citra 4')
```

Baris kode di atas digunakan untuk memperbaiki orientasi masing-masing potongan citra dengan melakukan rotasi dan flipping sesuai kebutuhan, kemudian menyusunnya kembali dalam tampilan 2x2 menggunakan Matplotlib sehingga tiap bagian citra dapat ditampilkan dengan posisi yang sudah benar.

```python
def translasi(image, geser_x=0, geser_y=0): 
    image = np.array(image) 
    h, w = image.shape 
    hasil = np.zeros((h, w), dtype=image.dtype) 
    
    for i in range(h): 
        for j in range(w): 
            new_i = i + geser_y 
            new_j = j + geser_x 
            if 0 <= new_i < h and 0 <= new_j < w: 
                hasil[new_i, new_j] = image[i, j] 
    return hasil 
```

Baris kode di atas mendefinisikan fungsi translasi yang digunakan untuk menggeser posisi citra berdasarkan nilai geser_x dan geser_y, dengan cara memindahkan setiap piksel ke koordinat baru menggunakan perulangan, lalu menyimpannya ke array hasil selama masih berada dalam batas ukuran citra.

```python
trans1 = translasi(citra1Benar, geser_x = 15, geser_y = 15)
trans2 = translasi(citra4Benar, geser_x = -15, geser_y = 15)
trans3 = translasi(citra2Benar, geser_x = 15, geser_y = -15)
trans4 = translasi(citra3Benar, geser_x = -15, geser_y = -15)

plt.figure(figsize=(7,7))

plt.subplot(2, 2, 1)
plt.imshow(trans1, cmap = 'gray')
plt.title('Citra 1')

plt.subplot(2, 2, 2)
plt.imshow(trans2, cmap = 'gray')
plt.title('Citra 2')

plt.subplot(2, 2, 3)
plt.imshow(trans3, cmap = 'gray')
plt.title('Citra 3')

plt.subplot(2, 2, 4)
plt.imshow(trans4, cmap = 'gray')
plt.title('Citra 4')
```

Baris kode di atas digunakan untuk menggeser masing-masing citra hasil perbaikan ke arah tertentu menggunakan fungsi translasi, baik ke kanan, kiri, atas, maupun bawah dengan kombinasi nilai positif dan negatif, kemudian menampilkannya dalam susunan 2x2 menggunakan Matplotlib untuk melihat hasil pergeseran setiap citra.

```python
def dilatasi(image, p1=0, p2=0, p3=0, p4=0): 
    h, w = image.shape 
    dilated = np.zeros((h * 2, w * 2)).astype(int) 
    for y in range(h * 2): 
        for x in range(w * 2): 
            dilated[y][x] = image[int(y / 2)][int(x / 2)] 
    sy, ey = max(0, p1), min(h * 2, h * 2 - p2) 
    sx, ex = max(0, p3), min(w * 2, w * 2 - p4) 
    return dilated[sy:ey, sx:ex]
```

Baris kode di atas mendefinisikan fungsi dilatasi yang digunakan untuk memperbesar citra dengan cara menggandakan setiap piksel menjadi dua kali lipat (zoom sederhana), kemudian memungkinkan pemotongan bagian tertentu dari hasil pembesaran menggunakan parameter p1, p2, p3, dan p4 sebagai batas area yang diambil.

```python
dil1 = dilatasi(trans1, 62, 0, 62, 0)
dil2 = dilatasi(trans2, 62, 0, 0, 62)
dil3 = dilatasi(trans3, 0, 62, 62, 0)
dil4 = dilatasi(trans4, 0, 62, 0, 62)

plt.figure(figsize=(7,7))

plt.subplot(2, 2, 1)
plt.imshow(dil1, cmap = 'gray')
plt.title('Citra 1')

plt.subplot(2, 2, 2)
plt.imshow(dil2, cmap = 'gray')
plt.title('Citra 2')

plt.subplot(2, 2, 3)
plt.imshow(dil3, cmap = 'gray')
plt.title('Citra 3')

plt.subplot(2, 2, 4)
plt.imshow(dil4, cmap = 'gray')
plt.title('Citra 4')
```

Baris kode di atas digunakan untuk memperbesar sekaligus memotong setiap citra hasil translasi menggunakan fungsi dilatasi dengan parameter yang berbeda-beda, sehingga tiap citra mengalami penyesuaian posisi dan ukuran, kemudian hasilnya ditampilkan dalam susunan 2x2 menggunakan Matplotlib untuk melihat perbandingan masing-masing citra.

```python
def merge_hori(image1, image2):
    image1 = np.array(image1)
    image2 = np.array(image2)

    tinggi = max(image1.shape[0], image2.shape[0])
    lebar = image1.shape[1] + image2.shape[1]
    
    hasil = np.zeros((tinggi, lebar)).astype(int)
    hasil [0:image1.shape[0], 0:image1.shape[1]] = image1
    hasil [0:image2.shape[0], image1.shape[1]:] = image2
    
    return hasil;

def merge_verti(image1, image2):
    image1 = np.array(image1)
    image2 = np.array(image2)

    tinggi = image1.shape[0] + image2.shape[0]
    lebar = max(image1.shape[1], image2.shape[1])
    
    hasil = np.zeros((tinggi, lebar)).astype(int)
    hasil [0:image1.shape[0], 0:image1.shape[1]] = image1
    hasil [image1.shape[0]:, 0:image2.shape[1]] = image2
    
    return hasil;
```

Baris kode di atas mendefinisikan dua fungsi untuk menggabungkan citra, yaitu merge_hori yang menggabungkan dua gambar secara horizontal (berdampingan kiri dan kanan) dengan menyesuaikan tinggi dan menjumlahkan lebar, serta merge_verti yang menggabungkan dua gambar secara vertikal (atas dan bawah) dengan menjumlahkan tinggi dan menyesuaikan lebar, di mana keduanya membuat array baru sebagai tempat hasil dan menyalin piksel dari masing-masing citra ke posisi yang sesuai.

```python
merge_atas = merge_hori(dil1, dil2)
merge_bawah = merge_hori(dil3, dil4)

plt.imshow(merge_atas, cmap='gray')
plt.show()

plt.imshow(merge_bawah, cmap='gray')
plt.show()
```

Baris kode di atas digunakan untuk menggabungkan citra hasil dilatasi menjadi dua bagian, yaitu bagian atas (merge_atas) dan bagian bawah (merge_bawah) dengan metode horizontal, kemudian masing-masing hasil penggabungan ditampilkan menggunakan Matplotlib dalam bentuk grayscale.

```python
citraMerge = merge_verti(merge_atas, merge_bawah)

plt.imshow(citraMerge, cmap = 'gray')
plt.show()
```

Baris kode di atas digunakan untuk menggabungkan dua citra hasil sebelumnya (merge_atas dan merge_bawah) secara vertikal menjadi satu citra utuh bernama kicau_merge, kemudian hasilnya ditampilkan menggunakan Matplotlib dalam bentuk grayscale.

```python
def tampilkan_histogram(citra, warna = "black"):
    histogram = np.zeros(256, dtype=int)

    for i in range(citra.shape[0]):
        for j in range(citra.shape[1]):
            nilai = int(citra[i, j])
            histogram[nilai] += 1
            
    plt.bar(range(256), histogram, width = 1, color = warna)
    plt.grid(True)

def normalisasi(image, warna = 'black'):
    image_min = image.min()
    image_max = image.max()

    image_normal = ((image - image_min) / (image_max - image_min) * 255)

    return image_normal
```

Baris kode di atas mendefinisikan dua fungsi, yaitu tampilkan_histogram yang digunakan untuk menghitung dan menampilkan histogram citra dengan menghitung jumlah setiap nilai intensitas piksel lalu divisualisasikan menggunakan grafik batang, serta normalisasi yang berfungsi untuk mengubah rentang nilai piksel citra menjadi 0–255 berdasarkan nilai minimum dan maksimum agar distribusi intensitas lebih merata.

```python
histogram = tampilkan_histogram(citraMerge, 'red')
```

Baris kode di atas digunakan untuk memanggil fungsi tampilkan_histogram pada citra citraMerge dengan warna merah, sehingga histogram distribusi intensitas piksel dari citra tersebut ditampilkan dalam bentuk grafik batang.

Baris kode di atas mendefinisikan fungsi bagi yang digunakan untuk membagi setiap nilai piksel pada citra menjadi setengah menggunakan operasi pembagian bilangan bulat (// 2), sehingga menghasilkan citra dengan intensitas yang lebih gelap.

Baris kode di atas digunakan untuk melakukan normalisasi pada citra hasil pembagian (citraBagi) sehingga rentang intensitasnya menjadi lebih merata, kemudian menampilkan perbandingan histogram sebelum dan sesudah normalisasi dalam satu tampilan menggunakan Matplotlib agar perubahan distribusi piksel dapat terlihat dengan jelas.

```python
kernel = np.array([
    [-1/10, -1/10, -1/10, -1/10],
    [-1/10,  4/10,  4/10, -1/10],
    [-1/10,  4/10,  4/10, -1/10],
    [-1/10, -1/10, -1/10, -1/10]
])

def zero_padding(citra, pad):
    height, width = citra.shape
    citra_pad = np.zeros((height + 2*pad, width + 2*pad), dtype=citra.dtype)
    citra_pad[pad:pad+height, pad:pad+width] = citra
    return citra_pad

def konvolusi(citra, kernel):
    kh, kw = kernel.shape
    pad_h = kh // 2
    pad_w = kw // 2
    citra_pad = zero_padding(citra, max(pad_h, pad_w))
    h, w = citra.shape
    hasil = np.zeros((h, w), dtype=np.float64)

    for i in range(h):
        for j in range(w):
            region = citra_pad[i:i+kh, j:j+kw]
            hasil[i, j] = np.sum(region * kernel)

    return hasil
```

Baris kode di atas mendefinisikan kernel untuk operasi konvolusi serta fungsi pendukung seperti zero_padding untuk menambahkan padding nol di sekitar citra agar ukuran tetap terjaga, dan konvolusi untuk menerapkan filter pada citra dengan cara mengalikan kernel pada setiap bagian citra lalu menjumlahkannya sehingga menghasilkan citra baru dengan efek tertentu seperti penajaman atau deteksi tepi.

```python
citraKonvo = konvolusi(citraNormal, kernel)

plt.figure(figsize=(15, 15))

plt.subplot(2, 2, 1)
plt.title('Sebelum Konvolusi')
plt.imshow(citraNormal, cmap = 'gray')

plt.subplot(2, 2, 2)
plt.title('Hasil Konvolusi')
plt.imshow(citraKonvo, cmap = 'gray')

plt.subplot(2, 2, 3)
plt.title('Histogram sebelum Konvolusi')
histo_kicau_normal = tampilkan_histogram(citraNormal, warna = 'blue')

plt.subplot(2, 2, 4)
plt.title('Histogram setelah Konvolusi')
histo_kicau_konvo = tampilkan_histogram(citraKonvo, warna = 'blue')
```

Baris kode di atas digunakan untuk menerapkan proses konvolusi pada citra hasil normalisasi (citraNormal) menggunakan kernel yang telah ditentukan sehingga menghasilkan citra baru (citraKonvo), kemudian menampilkan perbandingan citra sebelum dan sesudah konvolusi beserta histogram masing-masing dalam satu tampilan menggunakan Matplotlib agar dapat terlihat perubahan intensitas dan distribusi piksel akibat proses konvolusi.
